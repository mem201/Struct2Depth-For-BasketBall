```python
import tensorflow as tf
import os
import matplotlib.pyplot as plt
gfile=tf.gfile
```

```python
prediction_dir="/home/mice/Depth/struct2depth/refine"
model_ckpt="/home/mice/Depth/struct2depth/checkpoints/model-199160"
handle_motion="True"
size_constraint_weight="0"
data_dir="/home/mice/Depth/struct2depth"
triplet_list_file=str(data_dir)+"/triplet_list_file.txt"
```

```python
with gfile.FastGFile(triplet_list_file, 'r') as f:
    frames = f.readlines()
    frames = [line.rstrip() for line in frames]
    frames = [line for line in frames if len(line)]
```

```python
frames
```

```python
subfolders = [x.split(' ')[0] for x in frames]
frame_ids = [x.split(' ')[1] for x in frames]
```

```python
image_file_list = [ os.path.join(data_dir, subfolders[i], frame_ids[i] + '.' +'png') for i in range(len(frames))]
```

```python
image_file_list
```

```python
segment_file_list = [
        os.path.join(data_dir, subfolders[i], frame_ids[i] + '-fseg.' +
                     'png')
        for i in range(len(frames))
    ]
```

```python
segment_file_list
```

```python
file_lists = {}
file_lists['image_file_list'] = image_file_list
file_lists['segment_file_list'] = segment_file_list
```

```python
with tf.Session() as sess:
    image_paths_queue = tf.train.string_input_producer(file_lists['image_file_list'],num_epochs=1 )   
```

```python
img_reader = tf.WholeFileReader()
image_key, image_contents = img_reader.read(image_paths_queue)
```
